```python
from youtube_transcript_api import YouTubeTranscriptApi
from collections import namedtuple

import math
from itertools import accumulate
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd
```

```python
# function of full transript -> portion

def fullToPortion(full):
    
    lines = full.split(' ')
    
    segmentLen = 300

    length_to_split = math.floor(len(lines)/segmentLen)*[segmentLen] + [len(lines) - math.floor(len(lines)/segmentLen)*segmentLen]

    output = [lines[x - y: x] for x, y in zip(
              accumulate(length_to_split), length_to_split)]

    word_count_cig = []
    word_count_smok = []
    word_count_tobacco = []
    word_count_vap = []
    word_count_nicotine = []
    word_count_e_liquid = []
    word_count_e_juice = []
    word_count_all = []

    for i in range(len(output)):
        count_cig = 0
        count_smok = 0
        count_tobacco = 0
        count_vap = 0
        count_nicotine = 0
        count_e_liquid = 0
        count_e_juice = 0
        count_all = 0

        for k in output[i]:
            if 'cig' in k:
                count_cig += 1
                count_all += 1
            if 'smok' in k:
                count_smok += 1
                count_all += 1
            if 'tobacco' in k:
                count_tobacco += 1
                count_all += 1
            if 'vap' in k:
                count_vap += 1
                count_all += 1
            if 'nicotine' in k:
                count_nicotine += 1
                count_all += 1
            if 'e-liquid' in k or 'eliquid' in k:
                count_e_liquid += 1
                count_all += 1
            if 'e-juice' in k or 'ejuice' in k:
                count_e_juice += 1
                count_all += 1

        word_count_cig.append(count_cig)
        word_count_smok.append(count_smok)
        word_count_tobacco.append(count_tobacco)
        word_count_vap.append(count_vap)
        word_count_nicotine.append(count_nicotine)
        word_count_e_liquid.append(count_e_liquid)
        word_count_e_juice.append(count_e_juice)
        word_count_all.append(count_all)


    # x axis values
    x = len(lines)
    # corresponding y axis values
    y = word_count_all

    total_word = 0

    for i in word_count_all:
        total_word += i

    ls_peak = []

    for i in range(len(word_count_all)):
        if i >= 2:
            if word_count_all[i] + word_count_all[i - 1] + word_count_all[i - 2] > 0.01 * total_word:
                ls_peak.append(i)

    portion_transcript = ''

    for n in ls_peak:
        for j in output[n]:
            portion_transcript += '{} '.format(j)
            
    return portion_transcript
```

```python

transcript = ''

with open('id_state_date.csv','r') as csvinput:
    with open('id_state_date_portion.csv', 'w') as csvoutput:
        writer = csv.writer(csvoutput, lineterminator='\n')
        reader = csv.reader(csvinput)

        all = []
        row = next(reader)
        row.append('transcript_portion')
        all.append(row)

        for row in reader:
            try:    
                ls = YouTubeTranscriptApi.get_transcript(row[0])
                for item in ls:
                    transcript += item["text"]
                    transcript += ' '
                row.append(fullToPortion(transcript))
                all.append(row)
                transcript = ""
            except:
                row.append('')
                all.append(row)
                print(row[0])
                pass
        
        writer.writerows(all)
csvinput.close()
csvoutput.close()
```
