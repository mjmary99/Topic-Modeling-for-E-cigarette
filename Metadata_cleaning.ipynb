```python
dropRow =[]
for index, row in df.iterrows():
    date_set = row["date"].split("-")
    if date_set[0][0]=="0":
        date_set[0] = date_set[0][1]
    if date_set[1][0]=="0":
        date_set[1] = date_set[1][1]
    dy = (int(date_set[0])*30) + (int(date_set[1]))+((int(date_set[2])-2007)*365)
    if dy >0:
        row["day"]= dy
    else:
        dropRow.append(index)
```

```python
df=df.drop(labels=dropRow, axis=0)

df
```
