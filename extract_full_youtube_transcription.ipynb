```python
with open('id_title_description_duplicate.csv', 'w', encoding='UTF8') as csvfile:

    fieldnames = ['id', 'title', 'description']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    
    for i in ls_namedtuple:
        writer.writerow({'id': i.id, 'title': i.title, 'description': i.description})
csvfile.close()
```

```python
# eliminate duplicate videos
with open('id_title_description_duplicate.csv', 'r') as in_file, open('id_title_description.csv', 'w') as out_file:
    seen = set() # set for fast O(1) amortized lookup
    for line in in_file:
        if line in seen: continue # skip duplicate

        seen.add(line)
        out_file.write(line)

in_file.close()
out_file.close()
```

```python
# add state column
ls_state = [ 'AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DC', 'DE', 'FL', 'GA',
           'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME',
           'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM',
           'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX',
           'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']

with open('id_title_description.csv','r') as csvinput:
    with open('id_title_description_state.csv', 'w') as csvoutput:
        writer = csv.writer(csvoutput, lineterminator='\n')
        reader = csv.reader(csvinput)

        all = []
        row = next(reader)
        row.append('state')
        all.append(row)

        for row in reader:
            if row[1][:2] in ls_state:                
                row.append(row[1][:2])   
                all.append(row)
            else:
                row.append('')
                all.append(row)
        
        writer.writerows(all)

csvinput.close()
csvoutput.close()
```

```python
transcript = ''

with open('id_state_date.csv','r') as csvinput:
    with open('id_state_date_full.csv', 'w') as csvoutput:
        writer = csv.writer(csvoutput, lineterminator='\n')
        reader = csv.reader(csvinput)

        all = []
        row = next(reader)
        row.append('transcript')
        all.append(row)

        for row in reader:
            try:    
                ls = YouTubeTranscriptApi.get_transcript(row[0])
                for item in ls:
                    transcript += item["text"]
                    transcript += ' '
                #print(transcript)
                row.append(transcript)   
                all.append(row)
                transcript = ""
            except:
                row.append('')
                all.append(row)
                print(row[0])
                pass
        
        writer.writerows(all)
csvinput.close()
csvoutput.close()
```
